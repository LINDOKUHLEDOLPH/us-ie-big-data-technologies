```python
import pandas as pd

# Load the CSV file into a Pandas dataframe
df = pd.read_csv('users_v.csv')  # Update the path if necessary

# Display the top 10 rows of the dataframe
df.head(10)
```

```python
# Find the age of the oldest and youngest person in the dataset
oldest_age = df['age'].max()   # Oldest person's age
youngest_age = df['age'].min()  # Youngest person's age

# Print both values
print("Oldest Age:", oldest_age, "Youngest Age:", youngest_age)
```

```python
# Generate descriptive statistics for the 'age' column
print(df['age'].describe()[['count', 'mean', 'std']])
```

```python
# Create a 'surname' column using a lambda function to extract the last word from the 'name' column
df['surname'] = df['name'].apply(lambda x: x.split()[-1])

# Display the first few rows to verify
df[['name', 'surname']].head()
```

```python
# Display the last 10 lines of the dataframe
df.tail(10)

```

```python
# Convert 'date_joined' to a datetime type
df['date_joined'] = pd.to_datetime(df['date_joined'])

# Display the first few rows to verify the conversion
df[['date_joined']].head()

```

```python
# Convert 'date_joined' to a datetime type and overwrite the original column
df['date_joined'] = pd.to_datetime(df['date_joined'])

# Display the first few rows to confirm the change
df[['date_joined']].head()
```
